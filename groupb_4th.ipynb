```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"Airquality1.csv")
```

```python
data.info()
```

```python
data=data.drop("Unnamed: 0",axis=1)
data
```

```python
data['Ozone']=data['Ozone'].fillna(data['Ozone'].mean())
data['Solar.R']=data['Solar.R'].fillna(data['Solar.R'].mean())
```

```python
data['Humidity']=data['Humidity'].fillna(data['Humidity'].mode()[0])
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['Humidity']=label.fit_transform(data['Humidity'])
data['Humidity'].unique()
```

```python
data['Humidity']
```

```python
plt.scatter(x = data["Wind"],y = data["Temp"])
plt.title("Wind Scatter plot")
plt.xlabel("Wind")
plt.ylabel("Temperature")
plt.show()
```

```python
sns.pairplot(data) 
```

```python
corr = data.corr()
sns.heatmap(corr,annot = True)
plt.title("Correlation Heatmap Of Airquality Data")
```

```python
sns.barplot(x=data["Month"],y=data["Temp"]).set(title="Month vs Temperature")
```

```python
h = data.iloc[1:21, 2]
v = data.iloc[1:21, 3]
plt.plot(h, label="Wind", marker="o", linestyle="dotted")
plt.plot(v, label="Temp", marker="o", linestyle="dashed")
plt.title("Line graph for Wind and Temp")
plt.ylabel("Value Range")
plt.xlabel("Days")
plt.legend()
```

```python
sns.boxplot(x=data["Month"],y=data["Temp"])
```

```python
h=data.iloc[:,0]
plt.hist(h,bins='auto')
plt.title('Histogram ')
plt.xlabel("Ozone")
```

```python
import matplotlib.pyplot as plt
labels='Ozone','Solar.R','Wind',"Temperature"
sizes=[data['Ozone'].mean(),data["Solar.R"].mean(),data['Wind'].mean(),data["Temp"].mean()]
colors=['red','cyan','green','yellow']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=14)
plt.title("Aairquality Factors")
```
