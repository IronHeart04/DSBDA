```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
file= pd.read_csv("Airquality1.csv")
file
```

```python
file["Ozone"]= file["Ozone"].fillna(file["Ozone"].mean())
file["Solar.R"]= file["Solar.R"].fillna(file["Solar.R"].mean())
```

```python
file
```

```python
file.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
file
```

```python
plt.pie
```

```python
labels=['Ozone','Solar.R','Wind','Temp']
```

```python
file.columns
```

```python
sizes=[file['Ozone'].mean(),file['Solar.R'].mean(),file['Wind'].mean(),file['Temp'].mean()]
```

```python
colors=['red','green','pink','yellow']
```

```python
explode=(0.1,0,0,0)
```

```python
plt.pie(sizes,explode=explode,labeldistance=1.1,labels=labels,
colors=colors,autopct='%0.1f%%')
plt.title('Average Data')
plt.show()
```

```python
def addlabels():
 for i in range(len(file.iloc[0:21,3])):
  plt.text(i,file.iloc[0:21,3][i],file.iloc[0:21,3][i],ha='center')
    
y=np.arange(len(file.iloc[0:21,3]))
plt.figure(figsize=(10,6))
addlabels()
plt.bar(x=y,height=file.iloc[0:21,3],tick_label=y,color="dodgerblue")
plt.title('Wind measure for 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Wind')
plt.show()
```

```python
 file.boxplot(column=['Ozone','Wind','Temp','Solar.R'],grid=True)
plt.title('Box-Plot')
plt.suptitle('Temperature')
plt.ylabel('range')
plt.savefig('plot3.png')
```

```python
plt.figure(figsize=(10,6))
plt.hist(x=file['Temp'],bins=20,color="Red",label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("count")
plt.plot()
```
